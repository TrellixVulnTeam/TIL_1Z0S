## 4.1 선형결합(일차결합) - Linear combination


```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

  # 이미지 파일 불러오기
u = Image.open('images/adore.PNG')
v = Image.open('images/ah.PNG')
w = Image.open('images/anger.PNG')
u = u.convert('L')
v = v.convert('L')
w = w.convert('L')
v = v.resize(u.size)  # 이미지 사이즈를 u의 사이즈와 같게 맞추기
w = w.resize(u.size)
  
# 이미지 파일을 np.asarray를 이용해 배열로 만들기
u = np.asarray(u, dtype='float32')
v = np.asarray(v, dtype='float32')
w = np.asarray(w, dtype='float32')
  
# 스칼라 (1/3)을 곱하여 선형결합 하기
lin_comb = (1/3) * (u + v + w)

plt.imshow(lin_comb, cmap='Greys_r')

```

```python
u = Image.open('rian.PNG')
v = Image.open('ra.JPG')
w = Image.open('rian_flower.JPG')
u = u.convert('L')
v = v.convert('L')
w = w.convert('L')
v = v.resize(u.size)  # 이미지 사이즈를 u의 사이즈와 같게 맞추기
w = w.resize(u.size)
  
# 이미지 파일을 np.asarray를 이용해 배열로 만들기
u = np.asarray(u, dtype='float32')
v = np.asarray(v, dtype='float32')
w = np.asarray(w, dtype='float32')
  
# 스칼라 (1/3)을 곱하여 선형결합 하기
lin_comb = (1/3)*u +(1/2)*v +(1/6)*w

plt.imshow(lin_comb, cmap='Greys_r')

```

```python
from mpl_toolkits.mplot3d import Axes3D
xx, yy = np.meshgrid(range(10), range(10))  
zz = 1.65*xx + 1.0*yy  # 1.65x + 1y = 1z
  
ax = plt.subplot(projection='3d')
ax.plot_surface(xx, yy, zz)
plt.show()

```
