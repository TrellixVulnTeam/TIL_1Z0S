# 대각 행렬



A∈Cm×m 에 대해 A=Q∗ΛQ 를 만족하는 유니타리 행렬 Q와 대각행렬 λ가 존재하면, 행렬 A는 유니터리 대각화 가능하다고 말한다.

가정에서 고유벡터를 필요로 했으니 λ i는 고유벡터 x i의 고유값을 내타낸다는 것을 어렵지 않게 짐작할 수 있다.

이러한 대각화가 가능하다면 행렬의 거듭제곱을 아주 쉽게 구할 수 있다는 장점이 있다.



A가 대각화는 가능하지만 거듭제곱하기엔 굉장히 복잡하게 생겼다고 생각해보자.

원래는 엄청난 양의 계산을 수행해야하지만, 
A=SΛS−1 꼴로 나타난다면 An=SΛnS−1로도 나타낼 수 있다.

대각행렬 Λ=diag(λ1,λ2,⋯,λm)의 거듭제곱은 Λn=diag(λn1,λn2,⋯,λnm) 이므로 계산이 획기적으로 줄어든다.

```python
import numpy as np
from sympy import *
from numpy import linalg
x, y, z, t = symbols('x y z t')
f, g, h = symbols('f, g, h', cls=Function)

init_printing()
%matplotlib inline 
```

```python
H=np.diag([-3, -2, 1])   # 대각선행렬: diagonal_matrix([d1, d2, d3])
K=np.identity(3)# 단위행렬: identity_matrix(n)
K_1=np.identity(3)*3
J=np.zeros(4).reshape(2,2)   # 영행렬: matrix(m, n, 0)

print(H)
print(K)
print(K_1)
print(J)
```

```python
A = np.array([[1, -2], [2, -3]])
w1, V1 = np.linalg.eig(A)

print(w1)
print(V1)
```

```python
B = np.array([[2, 3], [2, 1]])
w2, V2 = np.linalg.eig(B)

print(w2)
print(V2)
```

```python
C = np.array([[0, -1], [1, 0]])
w3, V3 = np.linalg.eig(C)

print(w3)
print(V3)
```
